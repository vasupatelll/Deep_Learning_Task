### Imports<a name=1></a>

```python
import numpy as np 
import matplotlib.pyplot as plt 
import pandas as pd 
import keras
import os 
import keras.backend as k
import tensorflow as tf
import itertools

from keras.datasets import cifar10
from keras.preprocessing import image
from keras.preprocessing.image import ImageDataGenerator
from keras.applications import imagenet_utils, DenseNet121
from keras.optimizers import SGD, Adam
from keras.metrics import categorical_crossentropy, confusion_metrics
from keras.models import Model, Sequential
from keras import regularizers
from keras.layers import Dense, Dropout, Flatten, Conv2D, MaxPooling2D, Activation, BatchNormalization
from keras.utils.np_utils import to_categorical
from keras.callbacks import EarlyStopping
from keras.losses import categorical_crossentropy
```

### Loading & Preprocessing Dataset<a name=2></a>
<hr>

```python
# Loading data into (X,y) and (X,y), (train) and (test), respectively.
(X_train, y_train), (X_test, y_test) = cifar10.load_data()

# Converting X train and test datatype from integer to float
X_train = X_train.astype("float32")
X_test = X_test.astype("float32")
```

```python
X_train = X_train[:1000]
y_train = y_train[:1000]

X_test = X_test[:500]
y_test = y_test[:500]
```

```python
# Preprocessing image by using Data Augmentation method
datagen = ImageDataGenerator(
    featurewise_center=True,
    featurewise_std_normalization=True,
    rotation_range=20,
    width_shift_range=0.2,
    height_shift_range=0.2,
    horizontal_flip=True,
    validation_split=0.2)
```

### Pretrained Densenet-121<a name = 3 ></a>
<hr>

```python
dens = DenseNet121()

dens.summary()
```

```python
densenet_121 = DenseNet121(include_top= False, input_shape= (32, 32, 3))
densenet_121.summary()
```

```python
densenet_121.inputs
```

```python
# Defining the model architecture 
output = densenet_121.output
# print(output)

# images are 3d so have to flatten them into a (1 x 784) vector
flatten = Flatten()(output)
# print(flatten)

# # adding a Dense layer of 64 neurons with relu
dense = Dense(64, activation= "relu")(flatten)

batch_normalize = BatchNormalization()(dense)

# #Implementing dropout regularization with p = 0.3
dropout = Dropout(0.3)(batch_normalize)

# # adding an output layer (with 10 possible outputs for the 10 digits we need to predict)
prediction_layer = Dense(10, activation="softmax")(dropout)

model = Model(inputs = densenet_121.inputs, outputs= prediction_layer)
model.summary()
```

### Optimizer <a name =4 ></a>
<hr>

```python
adam_optimizer = Adam(learning_rate= 0.001)

model.compile(optimizer= adam_optimizer,
             loss= "categorical_crossentropy",
             metrics= ["accuracy"])
```

### Early stopping <a name = 5 ></a>
<hr>

```python
early_stopping = EarlyStopping(monitor= "val_loss",
                              mode= "min", 
                              verbose= 1,
                              patience= 10)
epochs = 100 
```

```python
history = model.fit(datagen.flow(X_train, y_train, batch_size=32,
         subset='training'),
         validation_data=datagen.flow(X_train, y_train,
         batch_size=8, subset='validation'),
         steps_per_epoch=len(X_train) / 32, epochs=epochs)
```
